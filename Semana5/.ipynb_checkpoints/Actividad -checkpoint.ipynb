```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import axes3d
from random import seed
from random import random 
from scipy import special
from scipy.stats import norm 
```

```python
#guamos a crear un array con un tamaño masivo na' más para ver qué onda xD le asignamos una letra cualquiera 
t=np.random.rand(200000)
```

```python
fig, (ax5, ax7) = plt.subplots(2, 1, figsize=(5,8))
ax5.hist(t, 15,  cumulative = False),ax7.hist(t, 15, cumulative = True)
ax5.set_title('Frecuencia no acumulada datos random'),ax7.set_title('Frecuencia acumulada datos random')
ax5.grid()
ax7.grid()
```

```python
np.random.seed(5)
#Inverse Cumulative Distribution Function xD
def ert(x, y, z):
    return special.erfinv(2*x-1)*z*2**0.5+y
np.random.seed(5)
fig, (ax,ax1, ax3) = plt.subplots(3,1,figsize = (5,10))
ax.hist(ert(t, 70, 2), 50,cumulative = True),ax1.hist(ert(t, 70, 2), 50),ax3.hist(np.random.randn(1234567)*2+70, 50)
ax.set_title("Frecuencia Acumulada inversa de CDF "),ax1.set_title("Datos inversa de CDF "),plt.title("Datos random.randn")
ax.grid()
ax1.grid()
ax3.grid()
plt.show()
```

```python
#Función de densidad de probabilidad 
def pf(s, w):
    return (s**(w/2-1)*np.e**(-s/2))/(2**(w/2)*special.gamma(w/2))
#Cumulative Distribution Function
def cf(s, w):
    return special.gammainc(w/2, s/2)/special.gamma(w/2)
x=np.linspace(0.0001, 20, 123456)
fig, (ax0, ax1) = plt.subplots(1, 2,figsize = (20,6))
for i in range(1,8):
    ax0.plot(x, pf(x, i), label =i)
ax0.set_ylim(-0.04, 1),ax0.set_xlim(-0.4, 20)
ax0.legend(),ax0.set_title("Función gamma")
ax0.grid()
for i in range(1,8):
    ax1.plot(x, cf(x, i), label =i)
ax1.set_xlim(-0.4, 20),ax1.legend(),ax1.set_title("Función gamma incompleta baja")
ax1.grid()
plt.suptitle("Distribución Chi-cuadrada")
plt.show()
```

```python
np.random.seed(5)
fig=plt.subplots(1,figsize = (8,8))
x=np.random.uniform(-1,1,10000)
y=np.random.uniform(-1,1,10000)
a=np.zeros(np.size(x), dtype=bool)
b=np.sqrt(x**2+y**2)
for i in range(np.size(x)):
    if (max(abs(x[i]), abs(y[i]))>0) & (b[i] < 1):
        a[i] = True
    else:
        a[i] = False
x = x[a]
y = y[a]
plt.title("Puntos aleatorios uniformemente distribuídos en un círuclo")
plt.plot(x,y,'.')
plt.show()
```

```python
np.random.seed(5)
fig=plt.subplots(1,figsize = (8,8))
x=np.random.uniform(-1,1,10000)
y=np.random.uniform(-1,1,10000)
z=np.random.uniform(-1,1,10000)
a=np.zeros(np.size(x), dtype=bool)
b=np.sqrt(x**2+y**2+z**2)
for i in range(np.size(x)):
    if (max(abs(x[i]), abs(y[i]),abs(z[i]))>0) & (b[i] < 1):
        a[i]=True
    else:
        a[i]=False
x=x[a]
y=y[a]
z=z[a]
ax=plt.axes(projection='3d')
ax.plot(x,y,z,'.')
ax.set_title("Puntos aleatorios uniformemente distribuídos en un esfera")
plt.show()
```
