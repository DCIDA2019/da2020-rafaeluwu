Averiguo el año mínimo y máximo de toda la tabla, dado que hay años en los que faltan datos de ciertos países.

```python
print(h['Year'].max())
print(h['Year'].min())
```

Quito de la tabla todos los datos que no corresponden al año dado para elaborar una gráfica de cada año.

Comienzo para cada tabla y gráfica

```python
h2003=h2003.sort_values('Count')
h2003=h2003.nlargest(10,['Count'])
height = h2003.Count
bars = h2003.Countries
y_pos = np.arange(len(bars))
plt.barh(y_pos, height)# crear barras horizontales
plt.yticks(y_pos, bars) # nombre en el y-axis
plt.ylabel("Country")
plt.xlabel("Number of prisioners")
plt.title("Prision Report 2003")
plt.show()
h2003
```
