# Creación de la base de datos

Nota: A lo largo de la realización de éste programa, me di cuenta de que cada vez que se une una tabla a otra, o un csv a otro csv el número de datos que despliega es menor.

```python
import numpy as np
import pandas as pd
```

Primero abrimos todos los archivos .

```python
tmin19=pd.read_csv('data/2019TMin.csv',encoding='latin-1')
# # # # # # # #
tmed19=pd.read_csv('data/2019TMed.csv',encoding='latin-1')
# # # # # # # #
tmax19=pd.read_csv('data/2019TMax.csv',encoding='latin-1')
# # # # # # # #
lluv19=pd.read_csv('data/2019Lluv.csv',encoding='latin-1')
```

Unimos los datos de las tablas y creamos una última tabla.

```python
t=pd.merge(tmin19,tmed19)
t=pd.merge(t,tmax19)
t
```

Primero revisamos las columnas de lluv19 para ver que coincidan para que no haya problema al momento de unir ambos

```python
lluv19.columns
```

Vemos que por alguna razón cuando crearon las tablas, el nombre de las columnas en lluv19 está escrito en mayúscula y en las tablas de temperatura en minúscula. 
Que mala organización.
Tendremos que cambiar los nombres de longitud, latitud, estado y nombre.

```python
rlluv19=lluv19.rename(columns={'LON':'Lon','LAT':'Lat','EDO':'Edo','ENE':'Ene','NOMBRE':'Nombre'})
rlluv19.columns
```

Una vez que ambos coinciden, procedemos a unirlos.Pero antes valos a quitar columnas que tienen valores repetidos, como CLAVE_SIH y Est, además vamos a reorganizar las columnas para acomodarlas en el mismo tipo de datos correspondientes.

```python
rlluv19=rlluv19.drop(columns='CLAVE_SIH')
t=t.drop(columns='Est')
```

```python
m=pd.merge(rlluv19,t)
m=m[['Lon','Lat','Clave','Edo','Nombre','Ene','Tmin','Tmed','Tmax']]
m
```
